```python
#Muhammad Anas bin Shamsul SW0106060
#Sean Christopher Marian IS01083522



# ================= IMPORT LIBRARIES =================

import time  # Used to add delays for page loading
import pandas as pd  # Used to store and export scraped data

# Selenium modules for browser automation
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Automatically installs correct ChromeDriver version
from webdriver_manager.chrome import ChromeDriverManager


# ================= CONFIGURATION =================

# Target Lazada product URL
PRODUCT_URL = "https://www.lazada.com.my/products/pdp-i2036148601-s8083722524.html"

# Maximum number of review pages to scrape
MAX_PAGES = 10


# ================= START WEB DRIVER =================

# Configure Chrome browser options
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  # Open browser in full screen
options.add_argument("--disable-blink-features=AutomationControlled")  # Reduce bot detection

# Initialize Chrome driver
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

# Open the product page
driver.get(PRODUCT_URL)

print("Waiting for page load or captcha solving...")
time.sleep(15)  # Allow manual captcha solving if required

# List to store extracted reviews
reviews = []


# ================= SCROLL TO REVIEWS =================

# Scroll down to trigger review section loading
driver.execute_script("window.scrollTo(0, document.body.scrollHeight*0.6);")
time.sleep(5)  # Wait for reviews to load


# ================= SCRAPE REVIEW PAGES =================

for page in range(1, MAX_PAGES + 1):

    print(f"Scraping Page {page}")

    try:
        # Wait until review elements are visible
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".item"))
        )
    except:
        print("Reviews not found")
        break

    # Get all review elements on page
    items = driver.find_elements(By.CSS_SELECTOR, ".item")

    if not items:
        print("No review items found")
        break

    # Extract data from each review
    for item in items:

        try:
            name = item.find_element(By.CSS_SELECTOR, ".middle span").text
        except:
            name = None  # If reviewer name not found

        try:
            content = item.find_element(By.CSS_SELECTOR, ".content").text
        except:
            content = None  # If review text not found

        try:
            date = item.find_element(By.CSS_SELECTOR, ".title").text
        except:
            date = None  # If review date not found

        # Store review data in dictionary format
        reviews.append({
            "Reviewer Name": name,
            "Review Date": date,
            "Review Content": content
        })

    print(f"Total reviews collected: {len(reviews)}")

    # ================= GO TO NEXT PAGE =================

    try:
        # Locate next page button
        next_btn = driver.find_element(By.CSS_SELECTOR, ".next-pagination-item.next")

        # Stop if last page
        if "disabled" in next_btn.get_attribute("class"):
            print("Last page reached")
            break

        # Click next page using JavaScript
        driver.execute_script("arguments[0].click();", next_btn)
        time.sleep(4)  # Wait for next page to load

    except:
        print("Next button not found")
        break


# ================= SAVE TO CSV =================

# Convert list to DataFrame
df = pd.DataFrame(reviews)

# Export data to CSV file
df.to_csv("lazada_reviews.csv", index=False)

print("DONE — Saved", len(df), "reviews")

# Close browser
driver.quit()
```
