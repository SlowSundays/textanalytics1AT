```python
import requests
from bs4 import BeautifulSoup
import csv
import os

# Base URL
base_url = "https://quotes.toscrape.com"
current_page_url = "/page/1/"
tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))

einstein_quotes = []
life_quotes = []

print(f"Starting Scraper. Current Directory: {os.getcwd()}")

# Loop through all pages
page_count = 1
while current_page_url:
    print(f"Scraping Page {page_count}...", end=" ")
    
    response = requests.get(base_url + current_page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    quote_elements = soup.find_all('div', class_='quote')
    
    for quote_div in quote_elements:
        text = quote_div.find('span', class_='text').get_text()
        author = quote_div.find('small', class_='author').get_text()
        
        # Filter for Albert Einstein
        if author == "Albert Einstein":
            einstein_quotes.append([text])
            
        # Filter for "life"
        if "life" in text.lower():
            life_quotes.append([text, author, tags])

    # Find the "Next" button
    next_button = soup.find('li', class_='next')
    if next_button:
        current_page_url = next_button.find('a')['href']
        page_count += 1
        print("Done.")
    else:
        current_page_url = None
        print("End of website reached.")

# --- SAVE AND FORCE REFRESH ---

print(f"\nSaving {len(einstein_quotes)} Einstein quotes to CSV...")
with open('quotes_Einstein.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Quote'])
    writer.writerows(einstein_quotes)

print(f"Saving {len(life_quotes)} Life quotes to CSV...")
with open('life_quotes.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Quote', 'Author', 'Tags'])
    writer.writerows(life_quotes)

print("\n--- ALL DONE! ---")
```
